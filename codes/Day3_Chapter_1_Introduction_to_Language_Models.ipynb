# Phi-3

The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately (although that isn't always necessary).

```python
from transformers import  AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
```

Although we can now use the model and tokenizer directly, it's much easier to wrap it in a `pipeline` object:

### The tokenizer is in charge of splitting the input text into tokens before feeding it to the generative model. <br /> <br />   You can find the tokenizer and model on the Hugging Face site and only need the corresponding IDs to be passed. <br />  In this case, we use ‚Äúmicrosoft/Phi-3-mini-4k-instruct‚Äù as the main path to the model.

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=50,
    do_sample=False
)
```

The following parameters are worth mentioning:

- return_full_text :
 By setting this to False, the prompt will not be returned but merely the output of the model.
- max_new_tokens :
The maximum number of tokens the model will generate. By setting a limit, we prevent long and unwieldy output as some models might continue generating output until they reach their context window.
- do_sample :
Whether the model uses a sampling strategy to choose the next token. By setting this to False, the model will always select the next most probable token. In Chapter 6, we explore several sampling parameters that invoke some creativity in the model‚Äôs output. <br/>  <br /> 
To generate our first text, let‚Äôs instruct the model to tell a joke about chickens. To do so, <br />  we format the prompt in a list of dictionaries where each dictionary relates to an entity in the conversation. Our role is that of ‚Äúuser‚Äù and we use the ‚Äúcontent‚Äù key to define our prompt:


Finally, we create our prompt as a user and give it to the model:

```python
# The prompt (user input / query)
messages = [
    {"role": "user",
     "content": "Create a punchline on Bahubali movie"}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
